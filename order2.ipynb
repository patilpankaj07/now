```python
from mlxtend.frequent_patterns import association_rules,apriori
from mlxtend.preprocessing import TransactionEncoder
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Order2.csv',header=None)
df.head()
```

```python
data=[]
for index,row in df.iterrows():
    temp=[j for j in row if pd.notna(j)]
    data.append(temp)
data
```

```python
len(data)
```

```python
data[:10]
```

```python
te=TransactionEncoder()
x=te.fit_transform(data)
x
```

```python
df1=pd.DataFrame(x,columns=te.columns_)
df1.head()
```

```python
df1.shape
```

```python
freq_pat=apriori(df1,min_support=0.01,use_colnames=True)
freq_pat
```

```python
rules=association_rules(freq_pat,metric='confidence',min_threshold=0.01)
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules
```
