```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Order1.csv')
df.head()
```

```python
data=df.groupby('Member_number')['itemDescription'].agg(list).to_list()
data
```

```python
data[1:10]
```

```python
len(data)
```

```python
te=TransactionEncoder()
x=te.fit_transform(data)
x
```

```python
df1=pd.DataFrame(x,columns=te.columns_)
df1.head()
```

```python
df1.shape
```

```python
freq_pat=apriori(df1,min_support=0.07,use_colnames=True)
freq_pat
```

```python
rules=association_rules(freq_pat,metric='confidence',min_threshold=0.01)
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules
```

```python
rules[rules['antecedents'] == {'whole milk'}]['consequents']
```
