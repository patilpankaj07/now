```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
df=pd.read_csv('Oder3.csv')
df.head()
```

```python
data=df.groupby('TransactionNo')['Items'].agg(list).to_list()
data
```

```python
len(data)
```

```python
data[:10]
```

```python
te=TransactionEncoder()
x=te.fit_transform(data)
x
```

```python
df1=pd.DataFrame(x,columns=te.columns_)
df1
```

```python
freq_pat=apriori(df1,min_support=0.001,use_colnames=True)
freq_pat
```

```python
rules=association_rules(freq_pat,metric='confidence',min_threshold=0.1)
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules
```
